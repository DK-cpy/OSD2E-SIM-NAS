```python
from init import init
init()
from natsbenchsss.models import NATSBenchResult
from nats_bench import create
api = create('/home/satan/NATS-sss-v1_0-50262-simple', 'sss', fast_mode=True, verbose=False)
```

```python
for i in range(32768):
    item = NATSBenchResult.objects.get(pk=i)
    tmp = {
            "cifar10-valid":dict(),
            "cifar10":dict(),
            "cifar100":dict(),
            "ImageNet16-120":dict()
        }
    x = api.query_by_index(i, hp='01').state_dict()
    assert item.phenotype == x['arch_str']
    for j in tmp:
        for w in ["01", '12', '90']:
            tmp[j][f"hp{w}"] = []

            # y = api.get_more_info_orm()
            for k in range(int(w)):
                tmp[j][f"hp{w}"].append(api.get_more_info_orm(index=i,dataset=j,iepoch=k,hp=w))
    assert tmp['cifar10-valid'] == item.cifar10_valid
    assert tmp['cifar10'] == item.cifar10
    assert tmp['cifar100'] == item.cifar100
    assert tmp['ImageNet16-120'] == item.ImageNet16_120
    if i % 16 == 0:
        print(i)

```
