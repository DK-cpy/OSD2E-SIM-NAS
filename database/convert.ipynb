```python
from nasbench101.models import NASBench101Result
fidelity = ['4', '12', '36', '108']
```

```python
for __, i in enumerate(NASBench101Result.objects.all()):
    i.params = i.phenotype["trainable_parameters"]
    i.flops = i.phenotype['flops']
    _ = dict()
    for j in fidelity:
        _[f'epoch{j}'] = [v['final_test_accuracy'] for v in i.result[j]]
    i.final_test_accuracy = _
    _ = dict()
    for j in fidelity:
        _[f'epoch{j}'] = [v['final_validation_accuracy'] for v in i.result[j]]
    i.final_validation_accuracy = _
    i.save()
    if __ % 1024 == 0:
        print(__)
```

```python
print(NASBench201Result.objects.all().count())
for __, i in enumerate(NASBench201Result.objects.all()):
    for j in i.cost12:
        i.cost12[j] = {k: i.cost12[j][k] for k in para}
    for j in i.cost200:
        i.cost200[j] = {k: i.cost200[j][k] for k in para}
    for j in i.more_info:
        for k in i.more_info[j]:
            i.more_info[j][k] = i.more_info[j][k][-1]
    i.save()
    if __ % 1024 == 0:
        print(__)
```

```python
print(NASBench201Result.objects.all().count())
para = ['valid-accuracy', 'test-accuracy', 'seed']
for __, i in enumerate(NASBench201Result.objects.all()):
    for j in i.more_info:
        for k in i.more_info[j]:
            for w in range(len(i.more_info[j][k])):
                # print(i.more_info[j][k][w])
                if 'valid-accuracy' in i.more_info[j][k][w]:
                    i.more_info[j][k][w] = {y: i.more_info[j][k][w][y] for y in para}
                else:
                    i.more_info[j][k][w] = {y: i.more_info[j][k][w][y] for y in para[1:]}
    # i.save()
    i.save()
    if __ % 1024 == 0:
        print(__)
```
